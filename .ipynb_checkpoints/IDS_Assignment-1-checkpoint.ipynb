# 3. Data Preparation

## 3.1 Check for

* duplicate data
* missing data
* data inconsistencies


```python
##---------Type the code below this line------------------##
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
missing_data = df.isnull().sum()
print(missing_data)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))].sum()
print(outliers)
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Column: {column}, Unique Values: {unique_values}")
```

## 3.2 Apply techiniques
* to remove duplicate data
* to impute or remove missing data
* to remove data inconsistencies


```python
##---------Type the code below this line------------------##
df = df.drop_duplicates()
df = df.dropna()
outliers = df[(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))].dropna()
```

## 3.3 Encode categorical data

```python
##---------Type the code below this line------------------##
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame
label_encoder = LabelEncoder()

# Encode each categorical column
df['Month'] = label_encoder.fit_transform(df['Month'])
df['Occupation'] = label_encoder.fit_transform(df['Occupation'])
df['Type_of_Loan'] = label_encoder.fit_transform(df['Type_of_Loan'])
df['Credit_Mix'] = label_encoder.fit_transform(df['Credit_Mix'])
df['Age'] = label_encoder.fit_transform(df['Age'])
df['SSN'] = label_encoder.fit_transform(df['SSN'])
df['Monthly_Balance'] = label_encoder.fit_transform(df['Monthly_Balance'])
df['Payment_Behaviour'] = label_encoder.fit_transform(df['Payment_Behaviour'])
df['Credit_History_Age'] = label_encoder.fit_transform(df['Credit_History_Age'])
df['Amount_invested_monthly'] = label_encoder.fit_transform(df['Amount_invested_monthly'])
df['Payment_of_Min_Amount'] = label_encoder.fit_transform(df['Payment_of_Min_Amount'])
df['Outstanding_Debt'] = label_encoder.fit_transform(df['Outstanding_Debt'])
df['Num_of_Delayed_Payment'] = label_encoder.fit_transform(df['Num_of_Delayed_Payment'])
df['Num_of_Loan'] = label_encoder.fit_transform(df['Num_of_Loan'])
df['Annual_Income'] = label_encoder.fit_transform(df['Annual_Income'])
df['Changed_Credit_Limit'] = label_encoder.fit_transform(df['Changed_Credit_Limit'])
df['ID'] = label_encoder.fit_transform(df['ID'])
df['Customer_ID'] = label_encoder.fit_transform(df['Customer_ID'])
df['Name'] = label_encoder.fit_transform(df['Name'])


df['Name'].fillna('Unknown', inplace=True)

# Impute missing values in 'Monthly_Inhand_Salary' column with the mean
mean_salary = df['Monthly_Inhand_Salary'].mean()
df['Monthly_Inhand_Salary'].fillna(mean_salary, inplace=True)

# Impute missing values in 'Type_of_Loan' column with the mode
mode_loan = df['Type_of_Loan'].mode()[0]
df['Type_of_Loan'].fillna(mode_loan, inplace=True)

# Impute missing values in 'Num_of_Delayed_Payment' column with 0
df['Num_of_Delayed_Payment'].fillna(0, inplace=True)

# Impute missing values in 'Num_Credit_Inquiries' column with the median
median_inquiries = df['Num_Credit_Inquiries'].median()
df['Num_Credit_Inquiries'].fillna(median_inquiries, inplace=True)

# Impute missing values in 'Credit_History_Age' column with the forward fill method
df['Credit_History_Age'].fillna(method='ffill', inplace=True)

# Impute missing values in 'Amount_invested_monthly' column with the median
median_investment = df['Amount_invested_monthly'].median()
df['Amount_invested_monthly'].fillna(median_investment, inplace=True)

# Option 2: Removing rows with missing data
# Remove rows with missing values in 'Payment_Behaviour' column
df.dropna(subset=['Payment_Behaviour'], inplace=True)

# Remove rows with missing values in 'Monthly_Balance' column
df.dropna(subset=['Monthly_Balance'], inplace=True)

# Option 3: Drop columns with a large number of missing values
# Drop 'Name' column as it has a large number of missing values
df.dropna(subset=['Name'], inplace=True)
```

```python
##---------Type the code below this line------------------##
import numpy as np
import pandas as pd
df.drop(["ID","Customer_ID","Month","Age", "SSN","Occupation"],axis = 1,inplace = True)
df['Credit_Risk'] = ((df['Credit_Utilization_Ratio'] > 0.7) | (df['Num_of_Delayed_Payment'] > 2) | (df['Outstanding_Debt'] > 10000)).astype(int)
df['Likelihood_of_Delayed_Payments'] = ((df['Delay_from_due_date'] > 5) | (df['Num_of_Delayed_Payment'] > 2) | (df['Payment_of_Min_Amount'] == 'No')).astype(int)
df['Output_Final']=((df['Credit_Risk']==1)&(df['Likelihood_of_Delayed_Payments']==1)).astype(int)
print("$$$$")
print(df['Output_Final'])
df.drop(["Credit_Risk","Likelihood_of_Delayed_Payments"],axis = 1,inplace = True)
# # Separate features (X) and labels (y)
# # Replace 'Credit_Risk' and 'Likelihood_of_Delayed_Payments' with your actual target variables
X= df.drop('Output_Final',axis=1)
y_output = df['Output_Final']
X[X < 0] = np.nan

# Impute negative values with mean of respective column
X = X.fillna(X.mean())
rows_with_negative_values = X[X.lt(0).any(axis=1)]

# Display rows with negative values
print(rows_with_negative_values)
print(y_output)
```

## 5.1 Univariate Filters

#### Numerical and Categorical Data
* Identify top 5 significant features by evaluating each feature independently with respect to the target/other variable by exploring
1. Mutual Information (Information Gain)
2. Gini index
3. Gain Ratio
4. Chi-Squared test
5. Strenth of Association

(From the above 5 you are required to use only any <b>two</b>)



Score: 3 Marks

```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import chi2
import numpy as np

# 1. Mutual Information
mutual_info = mutual_info_classif(X, y_output)
feature_mi_scores = dict(zip(X.columns, mutual_info))

# Sort the features based on their mutual information scores
top_5_mi_features = sorted(feature_mi_scores.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 features by Mutual Information:")
for feature, score in top_5_mi_features:
    print(f"{feature}: {score}")
    
# 2. Chi-Squared Test
chi2_scores, p_values = chi2(X, y_output)
feature_chi2_scores = dict(zip(X.columns, chi2_scores))

# Sort the features based on their chi-squared scores
top_5_chi2_features = sorted(feature_chi2_scores.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop 5 features by Chi-Squared Test:")
for feature, score in top_5_chi2_features:
    print(f"{feature}: {score}")
```

## 6.1 ML technique 1 + Justification

```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
import numpy as np

# Select top 5 features based on mutual information
top_features = ['Num_of_Delayed_Payment', 'Payment_of_Min_Amount', 'Delay_from_due_date', 'Credit_Mix', 'Num_of_Loan']
X_top = X[top_features]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_top, y_output, test_size=0.2, random_state=42)

# Initialize and train Logistic Regression model
logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X_train, y_train)

# Predictions
logreg_pred = logreg_model.predict(X_test)

# Evaluate Logistic Regression model
print("Logistic Regression:")
print("Classification Report:")
print(classification_report(y_test, logreg_pred))
print("Confusion Matrix ")
print(confusion_matrix(y_test, logreg_pred))
print("Accuracy Score - Logistic Regression:", accuracy_score(y_test, logreg_pred))
```

## 6.2 ML technique 2 + Justification

```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
import numpy as np

top_features = ['Num_of_Delayed_Payment', 'Payment_of_Min_Amount', 'Delay_from_due_date', 'Credit_Mix', 'Num_of_Loan']
X_top = X[top_features]
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_top, y_output, test_size=0.2, random_state=42)

# Initialize and train Random Forest Classifier model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
rf_pred = rf_model.predict(X_test)

# Evaluate Random Forest Classifier model
print("\nRandom Forest Classifier:")
print("Classification Report:")
print(classification_report(y_test, rf_pred))
print("Confusion Matrix ")
print(confusion_matrix(y_test, rf_pred))
print("Accuracy Score - Random Forest Classifier:", accuracy_score(y_test, rf_pred))
```
